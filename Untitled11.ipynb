```python
mydataset=pd.read_csv("train.csv")
```

```python
from pycaret.regression import *
setup(data=mydataset,target='target',normalize=True,normalize_method="minmax",
   transformation = True,
      transformation_method = 'quantile',
      verbose = False)
```

```python
cm=compare_models()
```

```python
mydataset.corr()['target'].sort_values(ascending=False)
```

```python
lightgbmModel=create_model("lightgbm")
```

```python
mynewdata=pd.read_csv("test.csv")
```

```python
predictions=predict_model(lightgbmModel,data=mynewdata)
predictions
```

```python
dictionary=pd.DataFrame({'id':predictions['id'],'target':predictions['prediction_label']})
```

```python
dictionary.to_csv("Answer.csv",index=False)
```
